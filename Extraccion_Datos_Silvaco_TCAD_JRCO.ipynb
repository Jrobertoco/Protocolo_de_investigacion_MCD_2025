```python
#Codigo V1 para extracción de datos a partir de un archivo .log generado en el software Silvaco TCAD.
import pandas as pd
from google.colab import files
import io
import sys

print("--- 1. Subir archivo ---")
print("Por favor, selecciona tu archivo .log (Photodiode_ex01_VD_SPECTRALRESPONSE.log):")

# Variable para guardar el contenido del archivo
file_content = None

try:
    # Mostrar el diálogo para subir archivos
    uploaded = files.upload()

    if not uploaded:
        print("\nOperación cancelada. No se subió ningún archivo.")
    else:
        # Obtener el nombre y el contenido del primer archivo subido
        log_filename = list(uploaded.keys())[0]
        file_content = uploaded[log_filename].decode('utf-8')
        print(f"\nArchivo '{log_filename}' subido y leído con éxito.")

except Exception as e:
    print(f"Ocurrió un error durante la subida: {e}", file=sys.stderr)

# --- 2. Procesar el archivo y crear el DataFrame ---

if file_content:
    print("\n--- 2. Procesando datos ---")

    # Lista para guardar los datos que vamos a extraer
    data_list = []

    # Dividir el contenido del archivo en líneas
    lines = file_content.splitlines()

    # --- Lógica de filtrado ---
    # Nota: Basado en el archivo, las líneas de datos reales comienzan con 'd '.
    # Usaremos esto como el filtro principal, ya que es más preciso
    # que un rango de líneas fijo (33-104), que podría incluir
    # encabezados o líneas vacías por error.

    for line in lines:
        # Eliminar espacios en blanco al inicio y al final
        line_stripped = line.strip()

        # Si la línea comienza con 'd ', es una línea de datos
        if line_stripped.startswith('d '):
            # Separar la línea por espacios
            parts = line_stripped.split()
            data_list.append(parts)

    if not data_list:
        print("No se encontraron líneas de datos (que comiencen con 'd ') en el archivo.")
    else:
        # --- 3. Crear el DataFrame ---
        try:
            # Crear el DataFrame de pandas a partir de la lista de datos
            tcad_data = pd.DataFrame(data_list)

            # Opcional: Asignar nombres de columna genéricos (ej. col_0, col_1, ...)
            tcad_data.columns = [f'col_{i}' for i in range(len(tcad_data.columns))]

            print("\n--- 3. ¡DataFrame 'tcad_data' creado! ---")
            print(f"Dimensiones: {tcad_data.shape[0]} filas, {tcad_data.shape[1]} columnas")

            print("\nPrimeras 5 filas del DataFrame 'tcad_data':")
            print(tcad_data.head())

            print("\nÚltimas 5 filas del DataFrame 'tcad_data':")
            print(tcad_data.tail())


        except Exception as e:
            print(f"Error al crear el DataFrame: {e}", file=sys.stderr)
```
